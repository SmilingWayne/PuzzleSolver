```python
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.desired_capabilities import DesiredCapabilities
import time 


def process_and_output(target, results, puzzle_grid, size):
    with open(f"../assets/data/Buraitoraito/problems/{target}_{size}x{size}.txt", "w+") as f:
            f.write(f"{size} {size}\n")
            for i in range(size):
                f.write(f"{' '.join(puzzle_grid[i * size:(i + 1) * size])}\n")
    with open(f"../assets/data/Buraitoraito/solutions/{target}_{size}x{size}.txt", "w+") as f:
        f.write(f"{size} {size}\n")
        for i in range(size):
            f.write(f"{' '.join(results[i * size:(i + 1) * size])}\n")
    print(f"Success: {target}")
    


driver = webdriver.Chrome()
options = webdriver.ChromeOptions()
options.add_argument("--disable-gpu")  # 禁用 GPU 渲染
total_takes = 0
size = 10
desire_cap = DesiredCapabilities.CHROME
desire_cap["pageLoadStrategy"] = "none"  # 禁用页面加载策略
scales = [[6,"r46ey"],[7,"9wygq"],[8,"nm2jw"],[9,"zq05n"]]

for scale in scales[2:3]:
    size, target = scale[0], scale[1]
    url = f"https://en.gridpuzzle.com/bright-light/{target}"
    driver.get(url)
    print(f"Change to Scale {size}")
    total_takes = 0
    while total_takes < 30:
        try:
            # desire_cap = DesiredCapabilities.CHROME
            # desire_cap["pageLoadStrategy"] = "none"  # 禁用页面加载策略

            # target = "8zg54"
            # url = f"https://cn.gridpuzzle.com/bright-light/{target}"
            # driver.get(url)
            # wait = WebDriverWait(driver, 10)
            # element = wait.until(EC.presence_of_element_located((By.ID, "puzzle")))
            # 等待网页加载完成（可调整时间或使用显式等待）
            # 获取网页源代码
            page_number = url.split("/")[-1]  # 或者使用 urlparse
            target = page_number
            div_elements = driver.find_elements(By.CLASS_NAME, "g_cell")

            # 初始化结果列表
            results = []
            puzzle_grid = []

            # 遍历每个 div
            for idx, div in enumerate(div_elements):
                # 获取 div 中的文本内容
                text = div.text.strip()  # 去掉多余的空白
                # 如果文本为空，获取 data-a 属性的值；否则保留文本
                if not text:
                    text = div.get_attribute("data-a") or "-"  # 默认为 "·"
                # 将编号和处理后的文本加入结果
                
                results.append(text)
                if text in ["*", "."]:
                    puzzle_grid.append("-")
                else:
                    puzzle_grid.append(text)
                
            if len(results) == size * size:
                process_and_output(target, results, puzzle_grid, size)

            try:
                next_button = driver.find_element(By.ID, "btn_go_next")
                ActionChains(driver).move_to_element(next_button).click(next_button).perform()
                try:
                    confirm_button = driver.find_element(By.CLASS_NAME, "swal2-confirm")
                    confirm_button.click()
                    print("已点击弹窗确认按钮")
                except Exception as e:
                    print(f"弹窗确认按钮未找到或无法点击: {e}")
                    url = f"https://cn.gridpuzzle.com/bright-light/{target}"
                    driver.get(url)
                    # 
                    # cancel_elem = driver.find_elements(By.ID, "dismiss-button")
                    # ActionChains(driver).move_to_element(cancel_elem).click(cancel_elem).perform()
                    
                # 等待页面跳转完成
                time.sleep(2)
                # 获取新页面的 URL
                url = driver.current_url
                total_takes += 1
                continue
            except Exception as e:
                print(f"无法点击下一题按钮: {e}")

        except Exception as e:
            print(f"发生错误: {e}")
            break
driver.quit()

```
